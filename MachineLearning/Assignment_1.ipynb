$f(x) = \begin{cases}
    \frac{1}{3} & 0 < x < 3 \\
    0           & \text{otherwise}
    \end{cases}$

**ANSWER**
Since the distribution function is zero for all negative values of x, the cumulative distribution function is also zero for negative values of x. For x in range \[0,3\],
\begin{equation}
F(x) = \int_{0}^{x}{f(x)}dx \\
= \int_{0}^{x}{\frac{1}{3}}dx \\
= \frac{1}{3}x \bigg\rvert_{0}^{x} \\
= \frac{1}{3}x
\end{equation}

For $x\gt3$, the cumulative value is 1 and remains constant throughout. The cumulative distribution function (CDF) for $f(x)$ is,
$F(x) = \begin{cases}
    0 & x < 0 \\
    \frac{1}{3}x & 0 \leq x \leq 3 \\
    1           & x > 3
    \end{cases}$

For this question I will be using the Airbnb data for NYC for the year 2019. The data was obtauned from Kaggle and can be found [here](https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data). The data describes various listungs found on Airbnb for NYC. This data can be used to analyze how prices vary across neighborhoods, reviews, availability and so on. The data consists of 48895 listings and 16 columns. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('NYC_2019.csv')
print(f"Data dimension: {data.shape}")
data.head()
```

As canbe seen below the data has some missing values. Particularly, *last_review* and *reviews_per_month* have a large number of missing values. Almost 20% values are missing in these columns. As a result we will not be using these columns in the analysis.

```python
data.isnull().sum()
```

The neighborhoods in NYC are broadly divided into five groups, *Bronx*, *Brooklyn*, *Manhattan*, *Queens* and  *Staten Island*. The plot below shows the number of listings per neighborhood group. From the graph we can observe that *Brooklyn* and *Manhattan* have maximum listings whereas *Bronx* and *Staten Island* have very less listings.

```python
data_by_neighborhood = data.groupby('neighbourhood_group')
data_by_neighborhood.groups.keys()
plt.bar(data_by_neighborhood.groups.keys(), data_by_neighborhood.count().id)
plt.xlabel('Neighborhood Group')
plt.ylabel('Number of listings')
plt.title('Number of listings by Neighborhood Group')
plt.show()
```

In the graph below I have tried to visulaize the type of rooms based on neighborhood group. From the graph it is clear that shared room listigs are the lowest. In *Bronx*, *Brooklyn* and *Staten Island* the Entire home and private listings are similar, however in *Manhattan* and *Queens* it is not. In *Manhattan* entire homes are much more than private apartments

```python
data.room_type.unique()
keys = list(data.groupby(['neighbourhood_group', 'room_type']).groups.keys())

x=list(zip(keys, data.groupby(['neighbourhood_group', 'room_type']).count().id))

bars1 = []
bars2 = []
bars3 = []

for vals in range(5):
    bars1.append(x[3*vals][1])
    bars2.append(x[3*vals+1][1])
    bars3.append(x[3*vals+2][1])
          
barWidth = 0.25
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Entire Home')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='Private Room')
plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Shared Room')

plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island'])

plt.legend()
plt.show()
```

The next plot describes the price distribution based on latitude and longitude. Most of the listings are priced from $0-$150 and a few listings are costlier which might be premium housing listings. We can clearly see a midpriced green color cluster where majority listings are prices around $300.

```python
#plt.scatter(data.latitude, data.longitude, c=data.price, alpha=0.5)
#plt.legend()

viz = data[data.price<500].plot(kind='scatter', x='longitude', y='latitude', label='neighbourhood_group', c='price',
                  cmap=plt.get_cmap('jet'), colorbar=True, alpha=0.4, figsize=(10,8))
viz.legend()
```

Next we visualize if there is any relationship between minimum stay and price of the listing. Visually, there doesn't seem to be any relationhip between the variables. Most of the listings have a minimum stay requirement of less than 100. But the data conssists of values where the minimum stay required is 1200 days which is absurd. The data does not mention anything about this.

```python
data.plot(x='minimum_nights', y='price', kind='scatter', title='Price vs minimum stay', alpha=0.7)
```

In the last we try to see if there is any relationship between the number of reviews for a listing and its price. We can see that as the cheaper rooms have high number of reviews. This is probably due to the fact that cheap listings have more customers than pricy premium listings and hence more reviews as well.

```python
data.plot(x='number_of_reviews', y='price', kind='scatter', title='Price vs number of reviews', alpha=0.7)
```

Thus, we analyzed the listings using the neighborhood, type of house information. We also analyzed which factors affect the price of listings. We saw that number of reviews has correlation with teh price whereas visually we see that minimum stay does not seem to have any relarionship with price of the listing.
